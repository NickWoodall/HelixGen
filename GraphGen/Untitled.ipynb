```python
import pyrosetta
```

```python
import RelaxFastdesign as ros
import os
```

```python
direc= '../data/4H_dataset/models/'
```

```python
fL = os.listdir(direc)
```

```python
scoreList = []
nameList = []

for i,name in enumerate(fL):
    if name.endswith('.pdb'):
        name = name[:-4]
        pose = pyrosetta.pose_from_file(f'{direc}/{name}.pdb')
        score_pose = ros.scoreOnly(pose)
        scoreList.append([name, score_pose.sequence(), ros.get_pose_scores(score_pose), 0])
        nameList.append(name)
        
    if i%100==1:
        print(f'Total done is {i}')
```

```python
def save_scoreList_ref(direc,nameList,scoreList):

    with open(f'{direc}ref_nameList.pkl',"wb") as f:
        pickle.dump(nameList,f)

    with open(f'{direc}ref_scoreList.pkl',"wb") as f:
        pickle.dump(scoreList,f)
```

```python
import pickle
outDirec = '../Figures/GraphGen_Figure/output/'
save_scoreList_ref(outDirec,nameList,scoreList)
```
